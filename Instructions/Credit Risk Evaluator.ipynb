```python
import numpy as np
import pandas as pd
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
train_df
```

```python
 #Checking for missing values
test_df.isnull().sum()
```

```python
#drop the label to create the X data
X = train_df.drop('index', axis=1)
X
```

```python
X_dummies = pd.get_dummies(X)
print(X_dummies.columns)
X_dummies
```

```python
# Convert categorical data to numeric and separate target feature for training data
y_train = train_df['loan_status']
X_train = train_df.drop(columns = ["loan_status"])
X_train = pd.get_dummies(X_train)
X_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
y_test = test_df['loan_status']
X_test = test_df.drop(columns = ["loan_status"])
X_test = pd.get_dummies(X_test)
# X_dummies
X_test
```

```python
# add missing dummy variables to testing set
X_test = X_test.reindex(columns = X_train.columns, fill_value=0)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

clf = LogisticRegression().fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
log_scaled = LogisticRegression().fit(X_train_scaled, y_train)
print(f'Training Score: {log_scaled.score(X_train_scaled, y_train)}')
print(f'Testing Score: {log_scaled.score(X_test_scaled, y_test)}')
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
forest_scaled = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {forest_scaled.score(X_train, y_train)}')
print(f'Testing Score: {forest_scaled.score(X_test, y_test)}')
```

    I did not expect the data from Random Forest classifier to be affected by scaling because scaling is not necessary for random 
    forests due to it being a tree based model. As you can see, the scaled data was the same for both scaled and unscaled data. 

    The machine learning algorithms, random forest, and logistic regression are utilized for various classification and regression purposes. In this case, the Random Forest classifier seems to outperform Logistic Regression; however, the difference is not immense.

    Results:

    Logistic Regression on unscaled data: 0.5253083794130158
    Random Forest Classifier on unscaled data: 0.6180348787749894
